```python
monthly_orders_df=all_df.resample(rule='M', on='order_purchase_timestamp').agg({
    'order_id':'nunique',
    'price':'sum'
})

monthly_orders_df.index=monthly_orders_df.index.strftime('%d-%m-%Y')#Mengubah format order_purchase_timestamp menjadi tanggal-bulan-tahun

monthly_orders_df=monthly_orders_df.reset_index()
monthly_orders_df.rename(columns={
    'order_id':'order_count',
    'price':'revenue'
},inplace=True)
monthly_orders_df.head()
```

```python
monthly_orders_df=all_df.resample(rule='M', on='order_purchase_timestamp').agg({
    'order_id':'nunique',
    'price':'sum'
})

monthly_orders_df.index=monthly_orders_df.index.strftime('%Y-%m')#Mengubah format order_purchase_timestamp menjadi bulan-tahun

monthly_orders_df=monthly_orders_df.reset_index()
monthly_orders_df.rename(columns={
    'order_id':'order_count',
    'price':'revenue'
},inplace=True)
plt.figure(figsize=(12, 5))
plt.plot(monthly_orders_df["order_purchase_timestamp"], monthly_orders_df["order_count"], marker='.', linewidth=1, color="#72BCD4")
plt.title("Number of Orders per Month (2016-2018)", loc="center", fontsize=18)
plt.xticks(rotation=50,fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
rfm_bystate_df = all_df.groupby(by='customer_state', as_index=False).agg({
    'order_purchase_timestamp':'max', # mengambil tanggal order terakhir
    'order_id':'nunique', # menghitung jumlah order
    'price':'sum' # menghitung jumlah revenue yang dihasilkan
})

rfm_bystate_df.columns =['customer_state','max_order_timestamp','frequency','monetary']

# menghitung kapan terakhir pelanggan melakukan transaksi (hari)
rfm_bystate_df['max_order_timestamp'] = rfm_bystate_df['max_order_timestamp'].dt.date
recent_date = orders_df['order_purchase_timestamp'].dt.date.max()
rfm_bystate_df['recency'] = rfm_bystate_df['max_order_timestamp'].apply(lambda x:(recent_date-x).days)


rfm_bystate_df.drop('max_order_timestamp', axis=1, inplace=True)
rfm_bystate_df.head()
```

###selajutnya kita akan menggunakan teknik analisis lanjutan yang bernama RFM analysis. metode ini merupakan salah satu metode umum, yang digunakan untuk melakukan segmentasi pelanggan berdasarkan tiga parameter:
####1.Recency: parameter yang digunakan untuk melihat kapan terakhir seorang pelanggan melakukan transaksi.
####2.Frequency: parameter ini digunakan untuk mengidentifikasi seberapa sering seorang pelanggan melakukan transaksi.
####3.Monetary: parameter terakhir ini digunakan untuk mengidentifikasi seberapa besar revenue yang berasal dari pelanggan tersebut.

#####untuk study case pada data ini, kita akan menggunakan customer state untuk melihat tiga parameter RFM analysis.
